```python
#Question: Which types of schools award which type of degree?
#loc = ticker.MultipleLocator(base=1)

public = df['PREDDEG'][df['CONTROL']==1].value_counts().sort_index()
pvtnp = df['PREDDEG'][df['CONTROL']==2].value_counts().sort_index()
pvtfp = df['PREDDEG'][df['CONTROL']==3].value_counts().sort_index()
fig, (ax1) = plt.subplots(nrows=1, ncols=1, figsize=(10,6))
ax1.bar(pvtfp.index, pvtfp, width=0.7, align='center', color='#194b6a', label='Private For Profit')
ax1.bar(pvtnp.index, pvtnp, bottom = pvtfp, width=0.7, align='center', color='#3397d4', label='Private Non Profit')
ax1.bar(public.index, public, bottom = pvtnp+pvtfp, width=0.7, align='center', color='#c1dff2', label='Public')
ax1.set_title('Predominant Undergrad Degree Awarded by Type of College')
ax1.legend(loc='best')
ax1.set_xticklabels([0,'Not Classified','Certificates','Associates','Bachelors','Graduate'])

#Certificates is dominantly for profit colleges, while Bachelors/Graduate are dominantly Non Profits.
```

```python
#Question: How do each type of school compare in terms of salary outcome?
sch_types = ['Public', 'Private Non Profit', 'Private for Profit']
df['md_earn_wne_p6c'] = pd.to_numeric(df['md_earn_wne_p6'], errors=coerce)
df['mn_earn_wne_p6c'] = pd.to_numeric(df['mn_earn_wne_p6'], errors=coerce)

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(16,5))
sns.stripplot(x='CONTROL', y='md_earn_wne_p6c', data = df[df['PREDDEG']==3], alpha = 0.4, ax=ax1)
ax1.set_xticklabels(sch_types)
ax1.set_title('Median Salary 6 years by School Type for Colleges that Award Mostly Bachelors')
sns.stripplot(x='CONTROL', y='mn_earn_wne_p6c', data = df[df['PREDDEG']==3], alpha = 0.4, ax=ax2)
ax2.set_xticklabels(sch_types)
ax2.set_title('Mean Salary 6 Years by School Type for Colleges that Award Mostly Bachelors')
ax2.set_ylim(0,140000)

#For profits don't far too behind as compared to non profits and public schools and actually have higher minimums than non profits
#However, the max salary for for-profits is very low compared to the densities of public and non-profits. 
```

### Majors
We all know that certain majors are more highly valued in today's market than other majors. Because of this, certian majors get paid higher salaries after college. Thinking about what would affect the **median** salary within a school, you might hypothesize that schools with a large majority of one major would skew the median higher or lower assuming the people within this major get similar paying jobs. Majors from a school that have a low % yet a high median salary may be skewed up due to other majors in the school making it unclear if the major is the reason for the higher median salary. We can plot median earnings against % of majors for every major to see if we can find a correlation.
